## Uygulama 1


```
# Klavyeden girilen 2 sayıdan büyük olanı ekrana yazan geri bildirimli kodu yazın.
```



````python
def BuyukBul(sayi1,sayi2):
  if sayi1>sayi2:
    return sayi1
  elif sayi2>sayi1:
    return sayi2
  else:
    return 0

deger1  = int(input("Lütfen Birinci Sayıyı Giriniz"))
deger2  = int(input("Lütfen İkinci Sayıyı Giriniz"))

sonuc = BuyukBul(deger1,deger2)

if sonuc==0:
  print("Değerler Eşittir")
else:
  print("Büyük Olan Sayı: ",sonuc)
````

## Uygulama 3

```
# Faktöriyel Hesaplaması yapan fonksiyonu yazınız.
```



````python
def FaktoriyelHesapla(faktoriyel):
  sonuc=1
  for i in range(faktoriyel,1,-1):
    sonuc=i*sonuc
  return sonuc

print(FaktoriyelHesapla(3))
````

## Recursive (Özyineli) Fonksiyonlar

Recursive fonksiyonlar kendi kendini çağıran fonksiyonlardır. Kısaca bir döngü gibi düşünebiliriz burada dikkat etmemiz gereken şey sonsuz döngüye girmemesi için bir şart koşulu kullanmalıyız. 

## Uygulama 1



```
# Girilen indiste fibonacci değerini bulan özyinelemeli fonksiyonu yazınız.
```



````python
# 0,1,1,2,3,5 ......
def fibonacci(sayi):
    if sayi==0:
        return 0
    if sayi==1:
        return 1
    return fibonacci(sayi-1) + fibonacci(sayi-2)

sayi=int(input("Bir sayi girin :"))
print(fibonacci(sayi))
````
