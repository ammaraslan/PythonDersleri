## Encapsulation - Kapsülleme 

Bu konu aslında başlıktan da anlaşılacağı üzere tanımladığımız bazı özellik ve metotların dışarıdan erişiminin kısıtlanması anlamına geliyor. Bazı özellik ve metotların her kişi tarafından erişilmemesi gerekir. Örneğin öğretmenlerin maaşına zam yapmak her nesnenin yapacağı bir iş değil, sadece Yönetim kademesinin ulaşması gerekir bu metoda. Diğer nesnelerin ulaşamaması gerekmekte. İşte bu şekilde ulaşılamamasını istediğimiz özellik ve metotlara erişimi engelleyebiliyoruz.

Örnekte görüldüğü gibi öğretmenlerin maaşlarına ulaşılmasını değişkene __maaş ismini vererek engelledik. Herkesin ulaşabileceği değişkenleri genel değişken, herkesin ulaşmasını istemediğimiz değişkenleri de özel değişken olarak tanımlıyoruz. Bu örnekte maaş özelliğini bulamadığı için hata verdi. Yani aslında erişimi engelledik. Buna Encapsulation ismi verilmekte. Peki bu maaşa ulaşması gerekenler nasıl ulaşacak? Hemen nasıl ulaşabileceğinizi göstereyim




````python
class Personel:
  def __init__(self,adi,soyadi,maasi,personelTip):
    self.isim = adi
    self.soyisim = soyadi
    self.personelTipi = personelTip
    self.__maas=maasi
  def bilgiler(self):
    print(f"Merhaba {self.isim} {self.soyisim} Hoş Geldiniz Personel Tipiniz: {self.personelTipi} Maaşı: {self.__maas}")
  def GetMaas(self):
    return self.__maas
  def setMaas(self,maas):
    if maas>=0:
      self.__maas =maas
    else:
      self.__maas = maas * -1
````

````python
Ammar = Personel("Ammar","Aslan",5000,"Akademisyen")
Baris = Personel("Barış","Çelebi",5000,"Akademisyen")
Taha = Personel("Taha","Akalan",0,"Öğrenci")

#Taha.bilgiler()
#Baris.bilgiler()
'''kisiler = [Ammar,Baris,Taha]
Kadriye = Personel("Kadriye","Demir",0,"Öğrenci")
kisiler.append(Kadriye)
for kisi in kisiler:
  print(kisi)
'''
print(Ammar.GetMaas())
Ammar.setMaas(-4000)
print(Ammar.GetMaas())
````

## Overriding (Üzerine Yazma-Öncekini İptal Etme)

````python
class Personel():
  def __init__(self,Adi,Soyadi):
    self.adi = Adi
    self.soyadi = Soyadi
    print("Personel Sınıfının Kurucu Metotu Çalıştı")
  def MerhabaDe(self):
    print("Merhaba "+ self.adi +" "+self.soyadi + " Hoş Geldiniz.")
````

````python
class Akademisyenler(Personel):
  def __init__(self,Adi,Soyadi,Maas):
    print("Akademisyen Sınıfının Kurucu Metotu Çalıştı")
    super().__init__(Adi,Soyadi)
    self.maas = Maas
  def AkademisyenMerhabaDe(self):
    print("Merhaba Sayın Akademisyen "+self.adi+" " + self.soyadi + " Hoş Geldiniz. Maaşınız"+str(self.maas))
  def MerhabaDe(self):
    print("Akademisyen Merhaba "+ self.adi +" "+self.soyadi + " Hoş Geldiniz.")
````

````python
ali = Akademisyenler("Ali","Kemal",4000)
ali.MerhabaDe()
````

````python
#biirnci adım
from abc import ABC,abstractmethod
class Hayvan():
  @abstractmethod
  def yuru(): pass
  @abstractmethod
  def kos(): pass

class kus(Hayvan):
  def __init__(self):
    print("Kuş")
  pass
a=kus()
````

````python
#ikinci hata
from abc import ABC,abstractmethod
class Hayvan(ABC):
  @abstractmethod
  def yuru(): pass
  @abstractmethod
  def kos(): pass

class kus(Hayvan):
  def __init__(self):
    print("Kuş")
  pass
a=kus()
````

````python
from abc import ABC,abstractmethod
class Hayvan(ABC):
  @abstractmethod
  def yuru(): pass
  @abstractmethod
  def kos(): pass

class kus(Hayvan):
  def yuru(self): 
    print("yuruyor")
  def kos(self): 
    print("kosuyor")
  def __init__(self):
    print("Kuş")
  pass
a=kus()
a.yuru()
````

### Uygulama 2

````python
from abc import ABC, abstractmethod
class Insan(ABC):
  def __init__(self):
    print("Insan Sınıfının Kurucu Metodu Çalıştı")
  @abstractmethod
  def merhabaDe(): pass
  @abstractmethod
  def yuru(): pass

class Abdulkadir(Insan):
  def __init__(self):
    print("Abdulkadir Sınıfının Kurucu Metodu Çalıştı")
  def merhabaDe(self): 
    print("Merhaba Ben Abdulkadir")
  def yuru(self): 
    print("Yürüyorum")
abdulkadir = Abdulkadir()
abdulkadir.merhabaDe()
abdulkadir.yuru()
````
